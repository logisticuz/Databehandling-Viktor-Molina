# Uppgift 2d
Beräkna hur många barn under 18 år det finns i Sverige baserat på det här datasetet. Du får leta upp
statistik på hur stor totala befolkningen är i Sverige.

```python
# Definierar en funktion för att avgöra om en åldersgrupp är under 18 år.
def is_under_18(age_group):
    try:
        return int(age_group.split('-')[0]) < 18  # Delar strängen vid '-' och kontrollerar om första delen är under 18.
    except:
        return False  # Felkontroll

# Använder funktionen för att filtrera ut och summera befolkningen under 18 år.
barn_under_18 = vaccin_data[vaccin_data['Ålder'].apply(is_under_18)]['Befolkning'].sum()

# Beräknar den totala befolkningen representerad i datasetet.
total_befolkning = vaccin_data['Befolkning'].sum()

# Beräknar andelen barn under 18 år i den totala befolkningen.
andel_barn_under_18 = (barn_under_18 / total_befolkning) * 100

# Skriver ut antalet och andelen barn under 18 år.
print(f"Antal barn under 18 år: {barn_under_18}")
print(f"Andel barn under 18 år: {andel_barn_under_18:.2f}%")
```

# Uppgift 2e
Rita ett diagram som visar på åldersfördelningen av Sveriges befolkning.

```python
# Definierar en funktion för att spara Plotly-figurer som HTML-filer.
def save_plotly_fig(fig, filename, directory="../visualiseringar"):
    # Skapar målmappen om den inte redan finns.
    Path(directory).mkdir(parents=True, exist_ok=True)

    # Skapar hela sökvägen för den sparade filen.
    filepath = f"{directory}/{filename}.html"

    # Sparar figuren som en HTML-fil på angiven sökväg.
    fig.write_html(filepath)

    # Skriver ut bekräftelsemeddelandet med sökvägen till den sparade filen.
    print(f"Figure saved as {filepath}")
```

```python
# Skapar ett stapeldiagram för att visualisera befolkningens åldersfördelning.
fig = px.bar(vaccin_data, x='Ålder', y='Befolkning', title='Åldersfördelning av befolkningen')

# Anpassar diagrammets layout för bättre läsbarhet.
fig.update_layout(
    xaxis_title='Ålder',
    yaxis_title='Befolkning',
    hovermode='closest',
)

# Sparar och visar diagrammet.
save_plotly_fig(fig, "Åldersfördelning av befolkningen")
fig.show()
```

# Uppgift 2f
Rita stapeldiagram för andel med minst 1 dos per län, andel två doser per län och andel tre doser per
län

```python
# Förbereder data för visualisering av vaccinationsandel per dos och län.
melted_data = vaccin_data.melt(
    id_vars=['Län_namn'],
    value_vars=['Andel minst 1 dos', 'Andel minst 2 doser', 'Andel 3 doser', 'Andel 4 doser'],
    var_name='Dostyp',
    value_name='Andel'
)

# Skapar stapeldiagram för att visa vaccinationsandel per dos i län.
fig_stapel = px.bar(
    melted_data, 
    x='Län_namn', 
    y='Andel',
    color='Dostyp',
    barmode='stack',
    title='Vaccinationsandel per dos i län',
)

# Anpassar utseende och interaktivitet.
fig_stapel.update_traces(
    marker=dict(line=dict(width=0)),
    hovertemplate="<b>%{x}</b><br>Dostyp: %{variable}<br>Andel: %{y:.2f}%<br>",
    opacity=0.8
)

# Anpassar layout.
fig_stapel.update_layout(
    hovermode='x unified',
    xaxis_title='Län',
    yaxis_title='Andel',
)

# Sparar och visar diagrammet.
save_plotly_fig(fig_stapel, "Vaccinationsandel per dos i län")
fig_stapel.show()
```

# Uppgift 2f
Rita stapeldiagram för andel med minst 1 dos per län, andel två doser per län och andel tre doser per
län

```python
# Laddar in vaccinationsdata från Excel-filen som finns sparad i projektets filstruktur.
vaccine_data = pd.read_excel('..\\filer\\Folkhalsomyndigheten_Covid19_Vaccine.xlsx', sheet_name='Vaccinerade kommun och ålder')

# Filtrerar data för att endast inkludera information för Stockholms län och Västra Götalands län.
filtered_data = vaccine_data[vaccine_data['Län_namn'].isin(['Stockholms län', 'Västra Götalands län'])]

# Omvandlar data till långformat för att underlätta skapandet av stapeldiagram i Plotly. Detta steg gör det möjligt att jämföra olika doser i diagrammet.
melted_data = filtered_data.melt(
    id_vars=['Län_namn'], 
    value_vars=['Andel minst 1 dos', 'Andel minst 2 doser', 'Andel 3 doser', 'Andel 4 doser'],
    var_name='Dostyp', 
    value_name='Andel'
)

# Skapar stapeldiagrammet med hjälp av Plotly Express för att visualisera vaccinationsandel per dos i de utvalda länen.
fig = px.bar(melted_data, 
             x='Län_namn', 
             y='Andel', 
             color='Dostyp', 
             title='Vaccinationsandel per dos i olika län',
             labels={'Andel': 'Andel vaccinerade', 'Län_namn': 'Län'},
             barmode='group'
)

# Uppdaterar diagrammets layout för att förbättra läsbarheten och tillhandahålla relevant information när man för muspekaren över staplarna.
fig.update_layout(
    hovermode='closest',  
    xaxis_title='Län',
    yaxis_title='Andel vaccinerade',
    title='Vaccinationsandel per dos i olika län'
)
fig.update_traces(
    marker=dict(line=dict(width=0)), 
    hovertemplate="<b>%{x}</b><br>" +
                  "Dostyp: %{Dostyp}<br>" + 
                  "Andel: %{y:.2f}%<br>" +
                  "<extra></extra>",
    opacity=0.8
)

# Sparar det färdiga diagrammet som en HTML-fil för enkel delning och tillgång.
save_plotly_fig(fig, "Vaccinationsandel per dos i olika län")


fig.show()
```
