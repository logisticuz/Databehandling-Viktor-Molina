```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

covid_data_directory = 'C:\\Users\\vikto\\Documents\\GitHub\Databehandling-Viktor-Molina\\Lab\\filer\\Folkhalsomyndigheten_Covid19.xlsx'
vaccine_data_directory = 'C:\\Users\\vikto\\Documents\\GitHub\Databehandling-Viktor-Molina\Lab\\filer\\Folkhalsomyndigheten_Covid19_Vaccine.xlsx'


vaccin_data = pd.read_excel(vaccine_data_directory, sheet_name='Vaccinerade kommun och ålder')
```

```python
antal_län = vaccin_data['Län'].nunique()

print(f"Antal län: {antal_län}")
```

```python
antal_kommuner = vaccin_data['Kommun'].nunique()

print(f"Antal kommuner: {antal_kommuner}")
```

```python
total_befolkning = vaccin_data['Befolkning'].sum

print(f"Total befolkning: {total_befolkning}")
```

```python
def is_under_18(age_group):
    try:
        return int(age_group.split('-')[0]) < 18
    except:
        return False


barn_under_18 = vaccin_data[vaccin_data['Ålder'].apply(is_under_18)]['Befolkning'].sum()

total_befolkning = vaccin_data['Befolkning'].sum()
andel_barn_under_18 = (barn_under_18 / total_befolkning) * 100

print(f"Antal barn under 18 år: {barn_under_18}")
print(f"Andel barn under 18 år: {andel_barn_under_18:.2f}%")
```

```python
fig = px.bar(vaccin_data, x='Ålder', y='Befolkning')
fig.show()
```
